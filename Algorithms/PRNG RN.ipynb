## Considerations

Problem - create a road network for use in a virtual city in one hour. Results should be 'believable' in appearance.

Immediately I thought - you could probably just use a probabilistic graphical model:
- Sample nodes from some mixture of gaussians with a small constraint to make sure they aren't too close
- Then you sample edges from some gaussian process with a covariance giving you closeness node

Two things: this would probably end up being too random, and has a chance of violating the principle of planarity!

So the fundamental problem with road map generation in the first place is *planarity*. There are all sorts of criteria for planarity but basically we want to make sure we don't have edge crossings.

You could try something like generating random graphs and enforcing planarity with something like `Boyer-Myrvold` - which gets rid of the PQ tree and uses edge additions to compute a planar embedding. Could also use Path Addition Method from Tarjan, but these are crazy rabbit holes - there are ways of guaranteeing planarity at the outset.

#### IMPLEMENTATION 1 - Rapidly Exploring Random Trees

First approach:
- Rapidly exploring random trees.
- These are great at path finding, and don't end up with edge crossings!
- Really efficient too, I think they would be optimal for motion planning in complex/cluttered environments.
- Simple to implement at face value.
- Pick a random Sample, find NN, select an action from neighbour that heads towards random sample, create new Sample, add new sample, connect the two.
- Tune it to enforce 90 degree-ish angles perhaps?
- Weak in that it's a tree, not a graph! Not a realistic road network after all...

````python
import sys, random, math
from math import sqrt, cos, sin, atan2

#constants
EPSILON = 7.0 # max distance
NUMNODES = 1000

def dist(p1, p2):
    return sqrt((p1[0] - p2[0])*(p1[0] - p2[0]) + (p1[1] - p2[1])*(p1[1] - p2[1]))

def step_from_to(p1, p2):
    if dist(p1, p2) < EPSILON:
        return p2
    else:
        theta = atan2(p2[1] - p1[1], p2[0] - p1[0])
        return p1[0] + EPSILON*cos(theta), p1[1] + EPSILON*sin(theta)

nodes = []

nodes.append((0.0,0.0)) # Start in the corner

for i in range(NUMNODES):
    rand = random.random()*800.0, random.random()*800.0
    nn = nodes[0]
    for p in nodes:
        if dist(p,rand) < dist(nn,rand):
            nn = p

    newnode = step_from_to(nn,rand)
    nodes.append(newnode)
    print(newnode[0], ',', newnode[1], ', 0')
````

#### IMPLEMENTATION 3 Delaunay Triangulation + Voronoi Tesselations

- Voronoi Tesselations/Delaunay Triangulation - you guarantee planarity at the outset!
- Delaunay Triangulation: Given set of P discrete points, triangulation is such that no point P is in the circumcircle of any triangle in Triangulation P.

- Voronoi is the dual graph of the Delaunay Triangulation (Graph that has a vertex for each face of graph)
- Uniform planar space constructed of cells.
- Doesn't result in something.... "Gridlike" (western architecture?)
- Results in a complete graph too.

Issues:
- Not gridlike enough
- Way out nodes
- Hard turns on nodes
- Complete graph
- Our heightmap sucks

How to fix?
- Use some kind of electrostatic repulsion to normalise cell shapes? Minkowski sum of edges generated? Lloyd's relaxation?
- Collect nodes of RRT arterial road
- Generate points that don't conflict with those using electrostatic repulsion
- Generate voronoi
- Noise? Perlin at frequencies + exponential?
- Perlin outputs are complex beasts. Not good at describing real world watershed geomorphology. 
- Cellular noise?

````python
import numpy as np
from scipy.spatial import Delaunay
import scipy.spatial

points_in_2d = np.random.rand(20, 2)

def dot_product(u, v):
    return u[0]*v[0] + u[1]*v[1]

def cross2(u, v, w):
    """u x (v x w)"""
    return dot_product(u, w)*v - dot_product(u, v)*w

def norm_cross_squared(u, v):
    """|| u x v ||^2"""
    return square_vec(u)*square_vec(v) - dot_product(u, v)**2

def square_vec(u):
    return dot_product(u, u)


triangulation = Delaunay(points_in_2d)

p = triangulation.points[triangulation.simplices]

# Triangle vertices
A = p[:, 0, :].T
B = p[:, 1, :].T
C = p[:, 2, :].T

# See http://en.wikipedia.org/wiki/Circumscribed_circle#Circumscribed_circles_of_triangulationangles
a = A - C
b = B - C

circumscribed_circle = cross2(square_vec(a) * b - square_vec(b) * a, a, b) / (2*norm_cross_squared(a, b)) + C

# Grab the Voronoi edges
voronoi_cells = circumscribed_circle[:, triangulation.neighbors]
voronoi_cells[:, triangulation.neighbors == -1] = np.nan # edges at infinity, plotting those would need more work...

lines = []
lines.extend(zip(circumscribed_circle.T, voronoi_cells[:, :, 0].T))
lines.extend(zip(circumscribed_circle.T, voronoi_cells[:, :, 1].T))
lines.extend(zip(circumscribed_circle.T, voronoi_cells[:, :, 2].T))
````

````python
# Plot it
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection


lines_to_plot = LineCollection(lines, edgecolor='k')

# plt.hold(1)
fig = plt.figure()

plt.plot(points_in_2d[:,0], points_in_2d[:,1], '.')
plt.plot(circumscribed_circle[0], circumscribed_circle[1], 'o')
plt.gca().add_collection(lines_to_plot)
plt.axis('equal')
plt.xlim(-0.1, 1)
plt.ylim(-0.1, 1)


plt.show()
````
